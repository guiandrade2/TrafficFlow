```python
punctual_incidents_reshaped = np.zeros((351,1189,10))
punctual_incidents_reshaped.fill(-1)

closed_roads_reshaped = np.zeros((351,98,7))
closed_roads_reshaped.fill(-1)
```

```python
grouped_punctual_incidents = punctual_incidents.groupby(['Day of year']).agg(lambda x: list(x))
grouped_closed_roads = closed_roads.groupby(['Day of year']).agg(lambda x: list(x))
```

```python
len(grouped_punctual_incidents.columns)
```

```python
grouped_punctual_incidents
```

```python
len(grouped_closed_roads.columns)
```

```python
grouped_closed_roads
```

```python
for index,content in grouped_punctual_incidents.iterrows():
    real_index = index - 15
    real_size = len(content[0])
    
    for entrances in range(real_size):
        place_holder = []
        for i in range(10):
            place_holder.append(content[i][entrances])
        #print(real_index)  
        #print(entrances)
        punctual_incidents_reshaped[real_index][entrances] = np.array(place_holder)
        
    #print('Done')
```

```python
for index,content in grouped_closed_roads.iterrows():
    real_index = index - 15
    real_size = len(content[0])
    
    for entrances in range(real_size):
        place_holder = []
        for i in range(7):
            place_holder.append(content[i][entrances])
        #print(real_index)  
        #print(entrances)
        closed_roads_reshaped[real_index][entrances] = np.array(place_holder)
        
    #print('Done')
```
