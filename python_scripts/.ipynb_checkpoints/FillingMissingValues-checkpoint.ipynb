```python
missing_values = pd.read_csv("../featured_datasets/missing_values.csv")
traffic_dataset = pd.read_csv("../featured_datasets/traffic_dataset.csv")
missing_values.describe()
```

```python
traffic_dataset.describe()
```

```python
pd.options.mode.chained_assignment = None

for i in range(0,len(traffic_dataset)):
    if(pd.isnull(traffic_dataset['temperature'][i]) == True):
        traffic_dataset['temperature'][i] = missing_values['temperature'][traffic_dataset['Hour'][i]]
        
    if(pd.isnull(traffic_dataset['rain'][i]) == True):
        traffic_dataset['rain'][i] = missing_values['atmospheric_pressure'][traffic_dataset['Hour'][i]]
        
    if(pd.isnull(traffic_dataset['atmospheric_pressure'][i]) == True):
        traffic_dataset['atmospheric_pressure'][i] = missing_values['atmospheric_pressure'][traffic_dataset['Hour'][i]]
    
    if(pd.isnull(traffic_dataset['humidity'][i]) == True):
        traffic_dataset['humidity'][i] = missing_values['humidity'][traffic_dataset['Hour'][i]]
        
    if(pd.isnull(traffic_dataset['wind_speed'][i]) == True):
        traffic_dataset['wind_speed'][i] = missing_values['wind_speed'][traffic_dataset['Hour'][i]]
    
    if(pd.isnull(traffic_dataset['clouds'][i]) == True):
        traffic_dataset['clouds'][i] = missing_values['clouds'][traffic_dataset['Hour'][i]]
    
    if(pd.isnull(traffic_dataset['current_luminosity'][i]) == True):
        traffic_dataset['current_luminosity'][i] = missing_values['current_luminosity'][traffic_dataset['Hour'][i]]
```

```python
traffic_dataset.describe()
```

```python
traffic_dataset.to_csv(r'../featured_datasets/traffic_dataset_complete.csv',index=False)
```
