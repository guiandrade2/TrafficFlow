```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../featured_datasets/ClosedRoads.csv")
```

```python
data
```

```python
data = data.to_numpy()
```

```python
new_data = []
tmp = 0

for row in data:
    
    if(tmp == 0):
        previous_day = row[0]
        previous_hour = row[1]
        previous_from_road = row[2]
        previous_to_road = row[3]
        
        new_data.append(data[tmp])
    else:
        
        
        day = row[0]
        hour = row[1]
        from_road = row[2]
        to_road = row[3]
        
        if(previous_from_road == from_road and previous_to_road == to_road):
            #ve se tá tudo direito no dia
            if(previous_day == day):
                #check hora
                #caso esteja mais 1 tudo bem
                if(hour == previous_hour + 1):
                    #tudo bem
                    new_data.append(data[tmp])
                else:
                    #prenche até chegar à hora
                    while(previous_hour != hour):
                        l = data[tmp]
                        previous_hour += 1
                        l[1] = previous_hour
                        new_data.append(l)
            else:
               
                while(previous_day != day and previous_hour != hour):
            
                    if(previous_hour == 24):
                        #preenche
                        previous_hour = 0
                        previous_day += 1
                        l = data[tmp]
                        l[0] = previous_day
                        l[1] = previous_hour
                        new_data.append(l)
                    else:
                        #preenche
                        previous_hour +=1
                        l = data[tmp]
                        l[0] = previous_day
                        l[1] = previous_hour
                        new_data.append(l)
                
        tmp+=1
    
```
