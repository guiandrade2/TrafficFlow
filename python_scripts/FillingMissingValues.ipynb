```python
pd.options.mode.chained_assignment = None
from scipy import stats

#Esta função preenche a luminosidade com uma moda nos últimos 10 dias

def filling_current_luminosity():

    last_cl = [] #ultimos 10 luminosidades
    last_cd = [] #ultimos 10 dias
    
    for df in (road_1,road_2,road_3,road_4):
        
        for row in df.iterrows():
            
            r = row[0]
            cl = row[1][12]#luminosidade
            cd = row[1][0]#dia
            
            
            if(pd.isnull(cl)):
                
                last_5_cl = last_cl[-5:]
                mode = stats.mode(last_5_cl)[0][0]
                df['current_luminosity'][r] = mode
                
                if(cd not in last_cd):
                    last_cd.append(cd)
                    last_cd.append(mode)
                
            else:
                #se já tem deste dia avança, caso contrario mete a moda
                if(cd not in last_cd):
                    last_cd.append(cd)
                    last_cl.append(cl)
                
                
filling_current_luminosity()              
```

> Remover todas as linhas que contêm missing values no time ratio

```python
#quando encontra missing time ratio remove
def missing_time_ratio():
    
    indexes = []
    
    for row in traffic_dataset.iterrows():
    
        r = row[0]
        time_ratio = row[1][5]
            
        if(not pd.isnull(time_ratio)):
            indexes.append(r)
                
    return traffic_dataset.drop(indexes)

def removing_missing_time_ratio():
    
    indexes = []
    dfs = []
    
    for df in (road_1,road_2,road_3,road_4):
        
        for row in df.iterrows():
            
            r = row[0]
            time_ratio = row[1][5]
            
            if(pd.isnull(time_ratio)):
                indexes.append(r)
                
        
        dfs.append(df.drop(indexes))
        indexes = []
    
    return dfs[0],dfs[1],dfs[2],dfs[3]
        
road_1, road_2, road_3, road_4 = removing_missing_time_ratio()
traffic_dataset = missing_time_ratio()
```

> Interpolação linear em todos os outros atributos do weather

> Utilizo 3 horas antes do missing value, e 3 horas após encontrar algum valor após o missing value

```python
def fill_and_return(idx,x_axis,y_axis,missing_axis,df):
    
     
    miss_rain = np.round(np.interp(missing_axis,x_axis,y_axis))
    
    for row in df.iterrows():
            
        r = row[0]
        #caso este seja missing, todos os outros são
        value = row[1][idx]
            
        if(pd.isnull(value)):
            
            if(r in missing_axis):
                index = missing_axis.index(r)
                value = miss_rain[index]
                if(idx == 6):
                    df['rain'][r] = value
                elif(idx == 7):
                    df['temperature'][r] = value
                elif(idx == 8):
                    df['atmospheric_pressure'][r] = value
                elif(idx == 9):
                    df['humidity'][r] = value
                elif(idx == 10):
                    df['wind_speed'][r] = value
                elif(idx == 11):
                    df['clouds'][r] = value
                elif(idx == 13):
                    df['atmosphere'][r] = value
                
                
```

```python
def linear_interpolation_weather_features(df,max_hour_missing = 6):
    
    #primeiro verifico o espaço entre os missing, caso sejam mais que 6 horas de missing values, deixo como está
    #qd uma missing feature está presentes então todas as outras também estão, o que me facilita o processo
    
    x_axis = [] #valores presentes em x
    
    missing_axis = [] #valores missing do array
    temporay = []
    
    #todos os y values
    rain = []
    temperature = []
    atmospheric_pressure = []
    humidity = []
    wind_speed = []
    clouds = []
    atmosphere = []
    
    missing_row = 0
    
    for row in df.iterrows():
                
        #caso este seja missing, todos os outros são
        rain_value = row[1][6]
                
        if(pd.isnull(rain_value)):
            temporary.append(row[0])
            missing_row += 1 #1 hora missing
            
        else:    
            x_axis.append(row[0])
                
            #tem valor, existem missings antes? #Se sim, tem menos que 7 horas missing?
            #Se tiver interpolação linear, Caso contrário deixa missing
            if(missing_row != 0 and missing_row <= max_hour_missing):
                for i in temporary:
                    missing_axis.append(i)
                
            temporary = []
            missing_row = 0
                
            #apende dos valores
            rain.append(row[1][6])
            temperature.append(row[1][7])
            atmospheric_pressure.append(row[1][8])
            humidity.append(row[1][9])
            wind_speed.append(row[1][10])
            clouds.append(row[1][11])
            atmosphere.append(row[1][13])
            
    
    fill_and_return(6,x_axis,rain,missing_axis,df)
    fill_and_return(7,x_axis,temperature,missing_axis,df)
    fill_and_return(8,x_axis,atmospheric_pressure,missing_axis,df)
    fill_and_return(9,x_axis,humidity,missing_axis,df)
    fill_and_return(10,x_axis,wind_speed,missing_axis,df)
    fill_and_return(11,x_axis,clouds,missing_axis,df)
    fill_and_return(13,x_axis,atmosphere,missing_axis,df)
        
        
linear_interpolation_weather_features(road_1)
linear_interpolation_weather_features(road_2)
linear_interpolation_weather_features(road_3)
linear_interpolation_weather_features(road_4)
```

> Now i have to put them all together again

```python
# join by: `df1.col1 == df2.index`
#df = pd.DataFrame(data=numpy_data, index=["row1", "row2"], columns=["column1", "column2"])

r1 = road_1.to_numpy()
r2 = road_2.to_numpy()
r3 = road_3.to_numpy()
r4 = road_4.to_numpy()

final_dataset = np.zeros(shape=(r1.shape[0]*4,r1.shape[1]))

for i in range(r1.shape[0]):
    tmp = i*4
    final_dataset[tmp] = r1[i]
    final_dataset[tmp+1] = r2[i]
    final_dataset[tmp+2] = r3[i]
    final_dataset[tmp+3] = r4[i]
    

columns = ['Day of year','Hour','week_or_not','road_num','functional_road_class_desc','time_rate','rain','temperature','atmospheric_pressure','humidity','wind_speed','clouds','current_luminosity','atmosphere']
final_dataset = pd.DataFrame(data=final_dataset, columns=columns)
```
